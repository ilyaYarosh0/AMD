```python
mising_in_powers = heroes[~heroes['name'].isin(powers['hero_names'])]
mising_in_heroes = powers[~powers['hero_names'].isin(heroes['name'])]

heroes_only_list = mising_in_powers['name'].dropna().unique().tolist()
powers_only_list = mising_in_heroes['hero_names'].dropna().unique().tolist()
print(f"{heroes_only_list}\n {powers_only_list}")
```

```python
# Lab №22 Tidy data



heroes_weight = (
    heroes[heroes['Publisher'].isin(['DC Comics', 'Marvel Comics'])] # filter
    .groupby(['Gender', 'Publisher'])                                # group_by
    .agg(weight_mean=('Weight', 'mean'))                             # summarise(mean)
    .dropna()                                                        # drop_na
    .reset_index() 
)

wide_table = heroes_weight.pivot(
    index ='Gender',
    columns = 'Publisher',
    values = 'weight_mean'
)

long_table = powers.melt(
    id_vars =['hero_names'],
    var_name='power',
    value_name = 'has'
).sort_values(by='hero_names')

wide_table_powers = long_table.pivot(
    index = 'power',
    columns = 'hero_names',
    values = 'has'

)
print(f"""{heroes_weight}\n
{wide_table}\n\n
{long_table}\n\n
{wide_table_powers}
""")
```

```python
# Lab №23 Операции с несколькими колонками:



na_counts = heroes.drop(columns=['Gender']).isna().groupby(heroes['Gender'], dropna=False).sum().reset_index()
print("1. Количество NA по всем колонкам:\n", na_counts.head())

color_cols = [col for col in heroes.columns if col.endswith('color')]
color_na_counts = heroes[color_cols].isna().groupby(heroes['Gender'], dropna=False).sum().reset_index()
print("\n2. Количество NA в колонках *color:\n", color_na_counts)

char_cols = heroes.select_dtypes(include=['object']).columns.drop('Gender', errors='ignore')

def get_longest_string(series):
    s_dropped = series.dropna()
    if s_dropped.empty:
        return np.nan
    return s_dropped.loc[s_dropped.str.len().idxmax()]

longest_strings = heroes.groupby('Gender', dropna=False)[char_cols].agg(get_longest_string).reset_index()
print("\n3. Самые длинные строки:\n", longest_strings)

heroes_global = heroes.copy()
for col in ['Height', 'Weight']:
    mean_val = heroes_global[col].mean()
    
    new_values = np.where(heroes_global[col] > mean_val, "выше среднего", "ниже среднего").astype(object)
    
    heroes_global[col] = np.where(heroes_global[col].isna(), np.nan, new_values)

print("\n4. Глобальные средние (первые 5 строк):\n", heroes_global[['name', 'Height', 'Weight']].head())


heroes_group = heroes.copy()
for col in ['Height', 'Weight']:
    group_means = heroes_group.groupby('Gender', dropna=False)[col].transform('mean')
    gender_text = heroes_group['Gender'].fillna('<NA>')
    
    new_values = np.where(
        heroes_group[col] > group_means, 
        "выше среднего по " + gender_text, 
        "ниже среднего по " + gender_text
    ).astype(object)
    
    heroes_group[col] = np.where(heroes_group[col].isna(), np.nan, new_values)
heroes.to_csv("heroes_clean.csv", index=False)

print("\n5. Групповые средние (первые 5 строк):\n", heroes_group[['name', 'Gender', 'Height', 'Weight']].head())
```

```python
# Lab №24 Описательная статистика



height = heroes['Height'].dropna()

mean_height = height.mean()

trimmed_mean_height = stats.trim_mean(height, proportiontocut=0.05)

print(f"""
Среднее: {round(mean_height, 4)}
Усеченное среднее (5%): {round(trimmed_mean_height, 4)}
Медиана: {height.median()}
Стандартное отклонение: {height.std()}
межквартильный размах: {stats.iqr(height)}
Ассиметрия: {round(height.skew(), 6)}
Эксцесс: {round(height.kurt(), 4)}\n
Множественные статистики:
{height.describe()}
""")
```
