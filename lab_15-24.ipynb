```python
import pandas as pd
import numpy as np
from scipy import stats

heroes = pd.read_csv("heroes_information.csv", na_values=["-", "-99", "NA", ""])
powers = pd.read_csv("super_hero_powers.csv")
```

```python
# Lab №19 Создание колонок: 



heroes['height_m'] = heroes['Height']/ 100
heroes_height_m = heroes[['name', 'height_m']]

heroes['hair'] = np.where(heroes['Hair color'] == 'No Hair', 'Bold', 'Hairy')

print(f"""{heroes_height_m.head(10)}\n
{heroes[['name', 'Hair color', 'hair']].head(10)}
""")
```

```python
# Lab №20 Агрегация:



race_count = heroes['Race'].value_counts(dropna=False)
height_by_gender = heroes.groupby('Gender', dropna=False)['Height'].mean()

print(f"{race_count.head()}\n{height_by_gender}")
```

```python
# Lab №21 Соединение датафреймов: *_join



merged_df = heroes.merge(powers, left_on='name', right_on='hero_names', how='inner')

web_creators = merged_df[merged_df['Web Creation'] ==True]

print(web_creators.head(16))
```

```python
mising_in_powers = heroes[~heroes['name'].isin(powers['hero_names'])]
mising_in_heroes = powers[~powers['hero_names'].isin(heroes['name'])]

heroes_only_list = mising_in_powers['name'].dropna().unique().tolist()
powers_only_list = mising_in_heroes['hero_names'].dropna().unique().tolist()
print(f"{heroes_only_list}\n {powers_only_list}")
```
