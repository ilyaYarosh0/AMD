```python
import numpy as np
```

```python
np.seterr(divide='ignore')
p = np.arange(4,8)
q = np.arange(0,4)
elemented_wise = p+q
dif = p-q
division = p/q
squared = p**q
mult = p*10
print(f"{elemented_wise}\n {dif}\n {division}\n {squared}\n {mult}")
```

```python
arr = np.arange(1,51)
res = sum(1/(arr*(arr+1)))
print(f"{res:.7f}")
```

```python
arr = np.arange(0, 21)
squ = np.sum(1/(2 ** arr))
print(f"{squ:.6f}")
```

```python
arr = np.arange(1,29,3)
n = np.arange(0,10)
squ = 3**n
res = np.sum(arr/squ)
print(f"{res:.6f}")
```

```python
arr = np.arange(1,29,3)
n = np.arange(0,10)
squ = 3**n
res = arr/squ
mask_great = res > 0.5
count = np.sum(mask_great)
print(f"{count}")
```
