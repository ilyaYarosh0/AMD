```python
import numpy as np
import pandas as pd
import string
```

```python
def trim(vector, n=1):
    if n < 0:
        raise ValueError("n не может быть меньше нуля")

    if n*2 >= len(vector) and n>0:
        raise ValueError("n слишком большой")
    
    if n ==0:
        return vector
    return vector[n:-n] 


num_vector = list(range(1, 8))

# Создаем список букв (аналог letters в R)[cite: 2]
letters = list(string.ascii_lowercase)

# 1. Проверка работы по умолчанию (отрезает по 1 значению)[cite: 2]
print(f"trim(1:7): {trim(num_vector)}")
print(f"trim(letters):\n{trim(letters)}\n")

# 2. Проверка работы с параметром n=2[cite: 2]
print(f"trim(letters, n=2):\n{trim(letters, n=2)}\n")

# 3. Проверка работы с параметром n=0 (возврат исходного вектора)[cite: 2]
print(f"trim(letters, n=0):\n{trim(letters, n=0)}")

print("\n--- Проверка ошибок ---")
try:
    trim(list(range(1, 7)), 3)
except ValueError as e:
    print(f"Ошибка при n=3: {e}") # Ожидаем: n слишком большой![cite: 2]

try:
    trim(list(range(1, 7)), -1)
except ValueError as e:
    print(f"Ошибка при n=-1: {e}")
```

```python
list3 = {
    'a': np.arange(1,6),
    'b': np.arange(0,21),
    'c': np.arange(4,25),
    'd': np.arange(6,2,-1),
    'e': np.arange(6,26)
}

max_list3 ={k: v.max().item() for k, v in list3.items()}
sum_list3 = {k: v.sum().item() for k, v in list3.items()}
len_list3 = {k: v.size for k, v in list3.items()}

print(f"Максимумы: {max_list3}\n Суммы: {sum_list3}\n Длины: {len_list3}")
```

```python
def max_item(list_dict):
    max_key = max(list_dict, key=lambda k: len(list_dict[k]))
    return list_dict[max_key]

print(f"max_item(list3):\n{max_item(list3)}")
```

```python
list4 = [np.arange(1, 4), np.arange(3, 41), list3]

def recursive_len(obj):
    if isinstance(obj, dict):
        return {k: recursive_len(v) for k, v in obj.items()}
    elif isinstance(obj, list):
        return [recursive_len(v) for v in obj]
    else:
        return len(obj) 

print(f"Длина элементов list4 (аналог rapply):\n{recursive_len(list4)}")
```
