```python
alphabet_list = list(string.ascii_lowercase)
num_array = np.arange(1,6)
list1 = {
    'numbers' : num_array,
    'letters' : alphabet_list,
    'logic' : True

}
# Имитация R-команды ist[1]
# Создаем новый словарь, забирая ключ  и его содержимое
sliced_list_numbers = {'numbers': list1['numbers']}
sliced_list_numbers2 = list(list1.values())[0]
print(f"{list1}\n {sliced_list_numbers['numbers']}\n {sliced_list_numbers2}\n {list1['numbers'][0]}")
```
