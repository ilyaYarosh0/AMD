```python
import numpy as np

def get_element(arr,index,default = "N/A"):
    try:
        return arr[index]
    except IndexError:
        return default
```

```python
arr = np.arange(3,28,3)
print(arr[1], arr[4], arr[6])
print("Предпоследний элемент", arr[-2])
print("Все кроме последнего", arr[:-1])
```

```python
vec3 = np.array([3, 5, 2, 1, 8, 4, 9, 10, 3, 15, 1, 11])
print(f"второй элемент {vec3[1]}")
print(f"второй и пятый элемент в {vec3[[1,4]]}")
print(f"сотое значение вектора {get_element(vec3, 99)}")
print(f"все элементы кроме второго элемента {np.delete(vec3, 1)}")
print(f"все элементы кроме второго и пятого элемента {np.delete(vec3, [1,4])}")
print(f"последний элемент вектора {vec3[-1]}")
print(f"все значения кроме первого и последнего {np.delete(vec3, [0,-1])}")
print(f"все значения, которые больше 4 {vec3[vec3 > 4]}")
print(f"все значения, которые больше 4, но меньше 10 {vec3[(vec3 > 4) & (vec3<10)]}")
print(f"все значения, которые больше 4 {vec3**2}")
```
