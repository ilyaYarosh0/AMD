```python
import pandas as pd
import statsmodels.api as sm

mtcars = sm.datasets.get_rdataset("mtcars").data
```

```python
print(mtcars.info())
print(mtcars.head())
```

```python
print(f"{mtcars.iloc[2,3]}\n {mtcars.iloc[:6,:6]}")
```

```python
print(f"{mtcars['wt']}\n")
```

```python
print(f"{mtcars[['hp', 'mpg', 'cyl']]}\n")
```

```python
print(f"""{(mtcars['cyl']==4).sum()}\n 
{((mtcars['cyl']==4).sum())/(mtcars['cyl'].count())}
""")
```

```python
print(mtcars[mtcars['hp']>=100])
```

```python
print(f"""{mtcars[(mtcars['hp']>=100) & (mtcars['cyl']==4)]}\n
max mass car: {mtcars['wt'].max()}
min mass car: {mtcars['wt'].min()}\n
{mtcars.loc[mtcars['wt'].idxmin()]}
""")
```

```python
print(f"строчки датафрейма mtcars с автомобилями, масса которых ниже средней массы.\n {mtcars[mtcars['wt']< (mtcars['wt'].median())]}")
```
