```python
m1 =np.ones((4,4))
print(m1, "\n")
m1[1:3, 1:3] = 2
print(m1)
print(f"2 и 3 столбцы матрицы\n {m1[:, 1:3]}\n")

row_vec = m1[1, :]
col_vec = m1[:, 1]
print (row_vec == col_vec)
```

```python
m2 = np.arange(1,31).reshape((5,6), order="F")
m2_del = np.delete(m2,2,axis=0)
m2_del = np.delete(m2_del, 1, axis=1)
m2_multiplication = m2_del*100
m2_squ = m2**2
m2_squ_minus = m2_squ - m2
print(f"{m2}\n\n {m2_del}\n\n {m2_multiplication}\n\n {m2_squ}\n\n {m2_squ_minus}")
```

```python
col = np.arange(1,10).reshape(9,1)
row = np.arange(1,10)
multi_table = col*row
sub_matrix = multi_table[5:8, 2:7]
bool_matrix = np.where(multi_table >= 10, True, False)
multi_table2 = np.where(multi_table <10, 0, multi_table)
print(f"{multi_table}\n\n {sub_matrix}\n\n {bool_matrix}\n\n {multi_table2}")
```
